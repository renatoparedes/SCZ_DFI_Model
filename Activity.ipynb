```python
from skneuromsi.neural import Paredes2022
import numpy as np
from findpeaks import findpeaks
```

```python
soa = 72

model = Paredes2022(
    time_range=(0, 600),
    neurons=30,
    position_range=(0, 30),
    tau=(6.712e00, 2.356e01, 1.175e02),
)

lspq_res = model.run(
    auditory_intensity=2.425,
    visual_intensity=1.375,
    auditory_stim_n=2,
    visual_stim_n=1,
    auditory_duration=7,
    visual_duration=12,
    auditory_soa=soa,
    noise=False,
    cross_modal_weight=1.327e-02,
    feedback_weight=2.423e-01,
    feedforward_weight=2.461e00,
    lateral_excitation=1.6,
)

hspq_res = model.run(
    auditory_intensity=2.425,
    visual_intensity=1.375,
    auditory_stim_n=2,
    visual_stim_n=1,
    auditory_duration=7,
    visual_duration=12,
    auditory_soa=soa,
    noise=False,
    cross_modal_weight=1.327e-02,
    feedback_weight=2.423e-01,
    feedforward_weight=2.461e00,
    lateral_excitation=1.632e00,
    cross_modal_latency=16,
)
```

```python
import string

plt.style.use("seaborn-v0_8-paper")

# initializes figure and plots
fig, axs = plt.subplots(2, 2, figsize=(10, 5), sharey="row", dpi=600, sharex="all")


ax1 = plt.subplot(221)
ax1.plot(lspq_res.extra_["auditory_total_input"].mean(axis=1))
ax1.set_ylabel(r"Actividad auditiva ($\mu_{ua}$)", size=10, weight="bold")
ax1.set_title("Modelo L-SPQ", weight="bold", size=12)
ax1.set_xticks(np.arange(0, 65000, 5000))
ax1.set_xticklabels(np.arange(0, 650, 50))

ax2 = plt.subplot(222)
ax2.plot(hspq_res.extra_["auditory_total_input"].mean(axis=1))
ax2.set_title("Modelo H-SPQ", weight="bold", size=12)
ax2.set_ylim(
    0,
)

ax3 = plt.subplot(223)
ax3.plot(lspq_res.extra_["auditory_total_input"].std(axis=1))
ax3.set_ylabel(r"Actividad auditiva ($\sigma_{ua}$)", size=10, weight="bold")
ax3.set_xlabel("Tiempo (ms)", size=10, weight="bold")

ax4 = plt.subplot(224)
ax4.plot(hspq_res.extra_["auditory_total_input"].std(axis=1))
ax4.set_xlabel("Tiempo (ms)", size=10, weight="bold")
ax4.set_ylim(
    0,
)

plt.tight_layout()

fig.savefig("drafts/FIG_Extra_1.eps", bbox_inches="tight", format="eps")
fig.savefig("drafts/FIG_Extra_1.png", bbox_inches="tight")
```
